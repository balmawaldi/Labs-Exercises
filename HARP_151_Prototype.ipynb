```python
import requests
import json
from bs4 import BeautifulSoup
from lxml import etree
source = requests.get("https://en.wikipedia.org/wiki/Tree").text
soup = BeautifulSoup(source, "lxml")
section = soup.find("div", class_= "mw-references-wrap mw-references-columns")
```

```python
#https://stackoverflow.com/questions/55611273/python-beautifulsoup-extracting-the-text-right-after-a-particular-tag
citations = []
for i in section.find_all('cite'):
    if ('from the original on' in i.text):
        cite = i.findNext('cite')
        if cite is not None:
            citations.append(cite.text)
        else:
            cite = None
            
ct = 1
for i in section.find_all("li"):
    d = {}  
    globals()[f"cit_data_{ct}"] = d
    ct += 1

ct = 1
for i in citations:
    citey = i
    wow = citey.split(".")
    d = {"title": "", "author": "", "date": "", "pub": ""}
    try:
        d["title"] = wow[1]
    except: 
        d["title"] = None
    try:
        d["author"] = wow[0]
    except: 
        d["author"] = None
    try:
        d["date"] = wow[3]
    except: 
        d["date"] = None
    try:
        d["pub"] = wow[2]
    except: 
        d["pub"] = None
    globals()[f"cit_data_{ct}"] = d
    ct += 1
    
cit_data_2
```
