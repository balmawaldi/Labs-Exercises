```python
PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver = webdriver.Chrome(service = PATH)

driver.get("https://www.youtube.com/@ArielBissett/")

click_xpath = "//*[@id='tabsContent']/tp-yt-paper-tab[2]/div"
link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, click_xpath)))
link.click()

try:
    main = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "style-scope ytd-rich-grid-renderer")))
    
    for i in driver.find_elements(By.CLASS_NAME, "style-scope ytd-rich-item-renderer"):
        link = i.find_element(By.ID, "video-title-link").get_attribute("href")
        title = i.find_element(By.XPATH, ".//*[@id='video-title']").text
        views = i.find_element(By.XPATH, ".//*[@id='metadata-line']/span[1]").text
        date = i.find_element(By.XPATH, ".//*[@id='metadata-line']/span[2]").text
        print(link, title, views, date)
        
finally:
    driver.quit()
```

```python
PATH = Service("C:/Program Files (x86)/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

driver.get("https://www.binghamton-ny.gov/home")

gov = driver.find_element(By.XPATH, "//*[@id='dropdownrootitem3']/a")
gov_dept = driver.find_element(By.XPATH, "//*[@id='dropdownrootitem3']/div/div/ul[1]/li/a")

actions = ActionChains(driver)
actions.move_to_element(gov)
actions.click(gov)
actions.click(gov_dept)
actions.perform()

csv_file = open("GovJobs", "w", newline = "", encoding = "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Job", "Type", "Application Deadline", "Salary"])

try:
    main =  WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id='widget_4_33_127']/ul/li[16]/a"))
    )
    main.click()
   
    main2 =  WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id='leftNav_1038_0_145']/ul/li/ul/li[14]/ul/li/a"))
    )
    main2.click()
   
    for i in range(1,15):
        job = driver.find_element(By.XPATH, f"//*[@id='ColumnUserControl4']/div[2]/table/tbody/tr[{i}]/td[1]").text
        hrs = driver.find_element(By.XPATH, f"//*[@id='ColumnUserControl4']/div[2]/table/tbody/tr[{i}]/td[2]").text
        deadline = driver.find_element(By.XPATH, f"//*[@id='ColumnUserControl4']/div[2]/table/tbody/tr[{i}]/td[3]").text
        salary =  driver.find_element(By.XPATH, f"//*[@id='ColumnUserControl4']/div[2]/table/tbody/tr[{i}]/td[4]").text
        csv_writer.writerow([job, hrs, deadline, salary])
        
finally:
    driver.quit()
    
csv_file.close()
```
