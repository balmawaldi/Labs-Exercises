```python
#button countdown
#code from https://python-course.eu/tkinter/buttons-in-tkinter.php

counter = 0 
def counter_label(label):
    counter = 0
    def count():
        global counter #global variable declaration 
        counter += 1
        label.config(text=str(counter))
        label.after(1000, count) #after(parent, ms, function = None, *args)
    count()
 
 
root = Tk()
root.title("Counting Seconds")
label = Label(root, fg="dark green")
label.pack()
counter_label(label)
button = Button(root, text='Stop', width=25, command=root.destroy) #destroy(parent, command, widgetobject.destroy) 
button.pack()
root.mainloop()
```

1) GUI #1 (5pts)
  - Using the quotes API, build a GUI that provides a drop-down menu of people randomly chosen from the API
  - I want a user to be able to pick from that list and receive a quote that corresponds to that person 
  - Ideally, I want to be able to reset that menu & allow people to choose from a new list 

```python
from tkinter import *
import requests
root = Tk()

quote_dict = {}
def getQuote():
    global quote_dict
    url = "https://api.quotable.io/random"
    info = requests.get(url).json()
    quote = info["content"]
    author = info["author"]
    quote_dict[author] = quote

def Quotes():
    global keys
    global values
    for i in range(4):
        getQuote()
    keys = quote_dict.keys()
    values = quote_dict.values()
    return quote_dict, str(keys), str(values)

def chosen_option():
    choquo = value.get()
    result_label = Label(root, text = quote_dict[choquo])
    result_label.pack()
 
def button():
    global value
    value = StringVar()
    value.set("Select an Author.")
    Quotes()
    dropdown = OptionMenu(root, value, *keys)
    dropdown.pack()
    choose_button = Button(root, text = "Click to choose", command = chosen_option)
    choose_button.pack()
    new_button = Button(root, text = "Give me new authors!", command = button)
    new_button.pack()
    
    
button()
root.mainloop() 
```

```python
import requests
import json

lat = "42.0894945"
lon = "-75.9708028"
genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")

json_file = genius.json() 

forecast = json_file["properties"]["forecast"]
new_request = requests.get(forecast)
json = new_request.json()

days = json["properties"]["periods"]
weat = {}
for i in days:
    weat[i['name']] = f"{i['temperature']}{i['temperatureUnit']}"
```
